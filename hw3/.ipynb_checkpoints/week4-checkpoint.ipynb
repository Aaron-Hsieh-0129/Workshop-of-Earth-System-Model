```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from pathlib import Path
```

```python
DATA_PATH = Path("/work/aaron900129/taiesm_work/archive") / "F_2000_TAI.f09_f09.r11229002.week4.seaice" / "atm/hist"
# DATA_PATH = Path("/work/aaron900129/taiesm_work/archive") / "f09.F2000.Aaron_week4" / "atm/hist"

nc_files = list(DATA_PATH.glob("*h0*.nc"))
nc_files.sort()

print(nc_files)
file = nc_files[0]
```

```python
ncd = nc.Dataset(file)
```

```python
lat = ncd["lat"]
lon = ncd["lon"]
```

```python
# f, ax = plt.subplots(figsize=(10, 8), subplot_kw={"projection": ccrs.Orthographic(central_longitude=0.0, central_latitude=90)})
f, ax = plt.subplots(figsize=(10, 8), subplot_kw={"projection": ccrs.Orthographic(central_longitude=0.0, central_latitude=90)})
ax.coastlines()
# ax.stock_img()
ax.set_extent([-180, 180, 60, 90], crs=ccrs.PlateCarree())

PC = ax.pcolormesh(lon, lat, ncd["ICEFRAC"][0], cmap="Blues", vmax = 1, vmin=0.5, transform=ccrs.PlateCarree())
cbar = plt.colorbar(PC, shrink=0.9)

gls = ax.gridlines(draw_labels=True, crs=ccrs.PlateCarree(), lw=1, color="gray",
        y_inline=True, xlocs=range(-180,181,30), ylocs=range(0,91,5))
```
